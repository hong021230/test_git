## 음식 주문 프로그램

```python
menu = {'자장면': 7000, '짬뽕': 8000, '볶음밥': 7000,  #메뉴판
        '마파두부': 10000, '탕수육': 25000, '깐풍기': 30000}

menu_order = [] # 메뉴 주문서
menu_sum = 0 # 총비용
answer = '예'

while answer == '예':
    print('메뉴는 다음과 같습니다')
    print(menu.keys())
    input_menu = input('메뉴를 선택하세요: ')
    menu_order.append(input_menu)
    menu_sum += menu[input_menu]
    answer = input('추가로 주문하시겠습니까?(예 또는 아니오): ')

print(f'주문하신 메뉴는 {menu_order} 입니다.')
print(f'결제 금액은 {menu_sum}원 입니다.')
```

## 음식 주문 프로그램
### 기능 추가: [오늘의 메뉴, 수량, 배달비]

```python
# 기능 추가: 오늘의 추천 메뉴 추가

import random

menu = {'자장면': 7000, '짬뽕': 8000, '볶음밥': 7000,  #메뉴판
        '마파두부': 10000, '탕수육': 25000, '깐풍기': 30000}

menu_order = [] # 메뉴 주문서
menu_sum = 0 # 총비용
answer = '예'
today_recommend = random.choice(list(menu.keys()))

while answer == '예':
    print('메뉴는 다음과 같습니다')
    print(menu.keys())
    print(f'오늘의 추천 메뉴는 {today_recommend}입니다. (추천 메뉴 선택시 10% 할인)')
    input_menu = input('메뉴를 선택하세요: ')
    menu_order.append(input_menu)
    if input_menu == today_recommend:
        menu_sum += menu[input_menu]*0.9 #오늘의 메뉴 10% 할인
    else:
        menu_sum += menu[input_menu]
    answer = input('추가로 주문하시겠습니까?(예 또는 아니오): ')

print(f'주문하신 메뉴는 {menu_order} 입니다.')
print(f'결제 금액은 {menu_sum}원 입니다.')
```

```python
#기능 추가: 수량 추가

import random

 #메뉴판
menu = {'자장면': 7000, '짬뽕': 8000, '볶음밥': 7000,
        '마파두부': 10000, '탕수육': 25000, '깐풍기': 30000}

menu_order = [] # 메뉴 주문서
menu_sum = 0 # 총비용
answer = '예'
today_recommend = random.choice(list(menu.keys()))

while answer == '예':
    print('메뉴는 다음과 같습니다')
    print(menu.keys())
    print(f'오늘의 추천 메뉴는 {today_recommend}입니다. (추천 메뉴 선택시 10% 할인)')
    input_menu = input('메뉴를 선택하세요: ')
    input_num = int(input('수량을 입력하세요: '))
    menu_order.append([input_menu, input_num])
    if input_menu == today_recommend:
        menu_sum += menu[input_menu]*0.9*input_num #오늘의 메뉴 10% 할인
    else:
        menu_sum += menu[input_menu]*input_num
    answer = input('추가로 주문하시겠습니까?(예 또는 아니오): ')
    print()

print('주문하신 메뉴는 다음과 같습니다')
for i, j in menu_order:
    print(f'{i}: {j}개')
print(f'결제 금액은 {menu_sum}원 입니다.')
```

```python
# 기능 추가: 배달여부 확인 후 배달비 추가

delivery = input('배달을 하시겠습니까? (예 또는 아니오): ')
delivery_fee = 0
print()

import random

# 메뉴판
menu = {'자장면': 7000, '짬뽕': 8000, '볶음밥': 7000,
        '마파두부': 10000, '탕수육': 25000, '깐풍기': 30000}

menu_order = []  # 메뉴 주문서
menu_sum = 0  # 총비용
answer = '예'
today_recommend = random.choice(list(menu.keys()))

while answer == '예':
    print('메뉴는 다음과 같습니다')
    print(menu.keys())
    print(f'오늘의 추천 메뉴는 {today_recommend}입니다. (추천 메뉴 선택시 10% 할인)')
    input_menu = input('메뉴를 선택하세요: ')
    input_num = int(input('수량을 입력하세요: '))
    menu_order.append([input_menu, input_num])
    if input_menu == today_recommend:
        menu_sum += menu[input_menu]*0.9*input_num  # 오늘의 메뉴 10% 할인
    else:
        menu_sum += menu[input_menu]*input_num
    answer = input('추가로 주문하시겠습니까?(예 또는 아니오): ')
    print()

# 배달비 계산
if delivery == '예':
    if menu_sum < 30000:
        delivery_fee = 3000
    else:
        delivery_fee = 0

total_sum = menu_sum + delivery_fee

print('주문하신 메뉴는 다음과 같습니다')
for i, j in menu_order:
    print(f'{i}: {j}개')
if delivery == '예':
    print(f'배달비: {delivery_fee}원')
else:
    print('배달비: 없음 (직접 수령)')
print(f'결제 금액은 {total_sum}원 입니다.')
```

## 음식 주문 프로그램 UI

```python
# Colab 실행용: 주문 UI + 표 요약(배달비 로직 포함)
# 셀 하나로 실행됩니다.

# 0) (필요 시) 위젯 설치
# - Colab은 보통 내장되어 있으나, 없을 경우 자동 설치됩니다.
try:
    import ipywidgets as widgets
except ModuleNotFoundError:
    %pip -q install ipywidgets
    import ipywidgets as widgets

import random
import pandas as pd
from IPython.display import display, clear_output

# 1) 메뉴 & 추천메뉴
menu = {
    '자장면': 7000,
    '짬뽕': 8000,
    '볶음밥': 7000,
    '마파두부': 10000,
    '탕수육': 25000,
    '깐풍기': 30000,
}
today_recommend = random.choice(list(menu.keys()))  # 오늘의 추천 (10% 할인)

# 2) 상태(장바구니)
order_rows = []  # {'메뉴', '단가', '수량', '할인(추천메뉴:bool)'}의 리스트

# 3) 유틸
def format_won(n: int) -> str:
    return f"{n:,}원"

def calc_totals(rows, delivery_on: bool):
    menu_sum = 0
    for r in rows:
        price = r['단가']
        qty = r['수량']
        if r['할인']:
            menu_sum += int(price * 0.9) * qty  # 추천메뉴 10% 할인
        else:
            menu_sum += price * qty
    # 배달비 로직
    delivery_fee = 0
    if delivery_on:
        delivery_fee = 0 if menu_sum >= 30000 else 3000
    total = menu_sum + delivery_fee
    return menu_sum, delivery_fee, total

def rows_to_df(rows):
    data = []
    for r in rows:
        price = r['단가']
        qty = r['수량']
        discounted = r['할인']
        line = int(price * 0.9) * qty if discounted else price * qty
        data.append({
            '메뉴': r['메뉴'],
            '단가': format_won(price),
            '수량': qty,
            '할인(추천메뉴)': '예' if discounted else '아니오',
            '금액': format_won(line),
        })
    return pd.DataFrame(data, columns=['메뉴', '단가', '수량', '할인(추천메뉴)', '금액'])

# 4) 위젯
title = widgets.HTML(f"<h3>중식 주문 (Colab UI)</h3>")
recommend_label = widgets.HTML(
    f"<b>오늘의 추천:</b> {today_recommend} (10% 할인)"
)

menu_dd = widgets.Dropdown(
    options=[(f"{k} — {format_won(v)}", k) for k, v in menu.items()],
    description='메뉴',
    layout={'width': '420px'}
)
qty_input = widgets.BoundedIntText(
    value=1, min=1, max=100, step=1, description='수량'
)
add_btn = widgets.Button(description='추가', button_style='primary')
reset_btn = widgets.Button(description='초기화', button_style='warning')

delivery_toggle = widgets.ToggleButtons(
    options=[('예', True), ('아니오', False)],
    value=False,
    description='배달',
)
tip = widgets.HTML("주문을 모두 추가한 후 배달 여부를 선택하세요.")

output_table = widgets.Output()
output_totals = widgets.HTML()

# 5) 렌더링
def render():
    df = rows_to_df(order_rows) if order_rows else pd.DataFrame(
        columns=['메뉴', '단가', '수량', '할인(추천메뉴)', '금액']
    )
    menu_sum, delivery_fee, total = calc_totals(order_rows, delivery_toggle.value)

    with output_table:
        clear_output(wait=True)
        if len(df) == 0:
            display(widgets.HTML("<i>아직 추가된 주문이 없습니다.</i>"))
        else:
            display(df)

    # 합계/배달비/총액 표시
    output_totals.value = (
        f"<b>주문 합계:</b> {format_won(menu_sum)}"
        f"<br><b>배달:</b> {'예(무료)' if delivery_toggle.value and delivery_fee == 0 else ('예' if delivery_toggle.value else '아니오')}"
        f"{'' if not delivery_toggle.value else f' — 배달비 {format_won(delivery_fee)}'}"
        f"<br><b>결제 금액:</b> {format_won(total)}"
    )

# 6) 이벤트
def on_add_clicked(_):
    item = menu_dd.value
    qty = qty_input.value
    order_rows.append({
        '메뉴': item,
        '단가': menu[item],
        '수량': qty,
        '할인': (item == today_recommend),
    })
    render()

def on_reset_clicked(_):
    order_rows.clear()
    qty_input.value = 1
    render()

def on_delivery_changed(change):
    if change['name'] == 'value':
        render()

add_btn.on_click(on_add_clicked)
reset_btn.on_click(on_reset_clicked)
delivery_toggle.observe(on_delivery_changed)

# 7) 배치 및 표시
controls = widgets.HBox([menu_dd, qty_input, add_btn, reset_btn])
summary_box = widgets.VBox([output_table, output_totals])

app = widgets.VBox([
    title,
    recommend_label,
    controls,
    widgets.HBox([delivery_toggle, tip]),
    summary_box
])

render()
display(app)
```
